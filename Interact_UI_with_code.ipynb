WorkspaceClient is one class inside the SDK that acts as the main entry point to the Databricks workspace.

```python
from databricks.sdk import WorkspaceClient
w = WorkspaceClient()
```

**1. Handling Clusters**

```python
for cluster in w.clusters.list():
    print(cluster.cluster_name, cluster.cluster_id, cluster.state)

# get specific cluster
cluster = w.cluser.get(cluster_id = "")

# creating cluster

# start a cluster
w.clusters.start(cluster_id = "")

# stop a cluster
w.clusters.stop(cluster_id = "")

# delete a cluster
w.clusters.delete(cluster_id = "")

# edit cluster
w.clusters.edit(cluster_id = "", new_settings = {"num_workers": 10})
```

**2. Jobs**

```python
# creating a job
job = w.job.create(
    name = "my-job",
    job_clusters=[
        JobCluster(
            job_cluster_key="main",
            new_cluster=ClusterSpec(
                spark_version="13.3.x-scala2.12",
                node_type_id="Standard_DS3_v2",
                num_workers=2
            )
        )
    ],
    tasks = [
        {
            "task_key": "my-task",
            "notebook_task": {
                "notebook_path": "path/to/notebook"
            }
        }
    ]
)
```

```python
# rujn a job
w.jobs.run_now(job_id = job.job_id, notebook_params={})

# get job status
w.jobs.get_run(run_id = 123456789)

# get job run output
output = w.jobs.get_run_output(run_id = 123456789)

# update, reset, cancle, delete job
```

**4. Secrets**
